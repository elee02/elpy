```python
import numpy as np
import matplotlib.pyplot as plt

# Define the two functions to convolve
def f1(t):
    return 2 * np.exp(-t) * (t >= 0)

def f2(t):
    return (t >= 0) * (t <= 1) * (t >= 0.5)

# Define the range of t values to evaluate the functions
t = np.linspace(-1, 3, 1000)

# Compute the convolution of the two functions
g = np.convolve(f1(t), f2(t), mode='same') / len(t)

# Plot the resulting function g(t)
plt.plot(t, g)
plt.xlabel('t')
plt.ylabel('g(t)')
plt.title('Convolution of Two Functions')
plt.show()
```
